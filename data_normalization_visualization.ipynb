### Normalizing the data collected at 4pm for all the conditions:
1. Indoors under shade
2. Indoors near window
3. Outdoors under shade
4. Outdoors directly under sun

```python
import pandas as pd

# The file name in the current working directory
file_name = '/content/drive/MyDrive/CSCE5280/data_4PM.xlsx'

# Load the Excel file
df = pd.read_excel(file_name)

# Columns to normalize (all except 'time_s')
columns_to_normalize = ['V_load_V', 'I_mA', 'P_mW', 'UVA', 'UVB', 'UVC']

# Create a copy of the original DataFrame to store the normalized values
df_normalized = df.copy()

# Perform Min-Max scaling for each column
# Min-Max Scaling formula: X_norm = (X - X_min) / (X_max - X_min)
for column in columns_to_normalize:
    min_val = df[column].min()
    max_val = df[column].max()
    # Check to prevent division by zero in case max_val == min_val
    if max_val != min_val:
        df_normalized[column] = (df[column] - min_val) / (max_val - min_val)
    else:
        # If max and min are the same, the normalized value is 0 (or simply keep the original value as it's a constant)
        # Since X_norm will be (const - const) / 0, we set it to 0 or 1, but for safety in this case, we'll set it to 0
        df_normalized[column] = 0.0

# Display the first few rows of the original and normalized DataFrames
print("--- Original Data Head ---")
print(df.head())

print("\n--- Normalized Data Head (Focus on 'time_s' being unchanged and others scaled) ---")
print(df_normalized.head())

# Save the normalized data to a new CSV file
output_file_name = '/content/drive/MyDrive/CSCE5280/normalized_data_4PM.xlsx'
df_normalized.to_excel(output_file_name, index=False)
print(f"\nNormalized data saved to {output_file_name}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Path to your Excel file
file_path = "/content/drive/MyDrive/CSCE5280/normalized_data_130PM.xlsx"

# Load the Excel file
df = pd.read_excel(file_path)

# Check if columns exist
print(df.columns)

# Extract relevant columns
voltage = df["V_load_V"]
current = df["I_mA"]

# Create a dummy timestamp (0,1,2,3,...)
df["timestamp"] = range(len(df))

# Compute trend line using rolling mean
voltage_trend = voltage.rolling(window=20, min_periods=1).mean()
current_trend = current.rolling(window=20, min_periods=1).mean()

# Plot
plt.figure(figsize=(12, 6))

plt.plot(df["timestamp"], voltage, label="V_load_V", alpha=0.4)
plt.plot(df["timestamp"], voltage_trend, label="V_load_V Trend", linewidth=2)

plt.plot(df["timestamp"], current, label="I_mA", alpha=0.4)
plt.plot(df["timestamp"], current_trend, label="I_mA Trend", linewidth=2)

plt.xlabel("Dummy Timestamp")
plt.ylabel("Values")
plt.title("Trend Line Plot: V_load_V and I_mA")
plt.legend()
plt.grid(True)
plt.show()
```
